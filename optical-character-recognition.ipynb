```python
import keras_ocr
import pandas as pd
import os

print("Current Working Directory:", os.getcwd())

def image_to_text(image_path):
    pipeline = keras_ocr.pipeline.Pipeline()

    # Read image
    images = [keras_ocr.tools.read(image_path)]

    # Generate (word, box) tuples
    prediction_groups = pipeline.recognize(images)

    # Extract text from predictions
    text = " ".join([word for word, box in prediction_groups[0]])
    return text

def save_text_to_csv(text, output_csv_path):
    try:
        text = text.encode('utf-8').decode('utf-8')

        # Create a DataFrame from the text
        data = {'Content': [text]}
        df = pd.DataFrame(data)
        
        # Save the DataFrame to a CSV file
        df.to_csv(output_csv_path, index=False, encoding='utf-8')
        print(f"CSV file saved at: {output_csv_path}")
    except Exception as e:
        print(f"Error occurred while saving the CSV file: {e}")

# input image
image_path = 'tm224644d1_ex99-1img005.jpg'
output_csv_path = 'output.csv'  # location of the file saved

text = image_to_text(image_path)
if text:
    save_text_to_csv(text, output_csv_path)
else:
    print("No text extracted from the image.")
```
